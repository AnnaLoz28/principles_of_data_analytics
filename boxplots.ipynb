## Bar Charts

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py

Bar charts are used for discrete type of data, not continuous.



### Example of a bar chart with individual bar colors

https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html#sphx-glr-gallery-lines-bars-and-markers-bar-colors-py

Plot fruits against their count, both discrete type of data.

X axis is discrete. The width of the bars is by default 0.8 (has no meaning) and bars are equally-spaced. 

The height of the bars, displayed on the y axis,  is instead important, as it indicates the count of the elements displayed on the x axis. 

```python
fig, ax = plt.subplots(); #create a figure and a set of subplots (even if we create just one plot now)

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange'] #list of labels of each bar. The underscore before 'red' is to make the label invisible.
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange'] #list of colors of each bar. tab: is a prefix to use the Tableau colors.

ax.bar(fruits, counts, label=bar_labels, color=bar_colors) #set the labels and colors of the bars

ax.set_ylabel('fruit supply') #set the y-axis label
ax.set_title('Fruit supply by kind and color') #set the title of the plot
ax.legend(title='Fruit color') #set the legend title
plt.show() #display the plot
```

```python
# simpler version of the above plot
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
plt.bar(fruits, counts);
```

## Histograms
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib.axes.Axes.hist

Histograms are used for continuous type of dat, where each value tends to be unique and trends can be observed. 




### Create data for analysis using NumPy Random Generator:

https://numpy.org/doc/stable/reference/random/generator.html

https://www.w3schools.com/python/numpy/numpy_random.asp

We want to generate a set of random normal numbers (normally distributed values).


```python
rng = np.random.default_rng(12345) #create a random number generator
group1 = rng.normal(75, 10, 30) #generate 30 random numbers with a normal distribution with mean 75 and standard deviation 10
print(group1)
```

Histograms can be used with continuous values and show trends, instead of single observations (like bar charts do).

```python
plt.hist(group1, edgecolor = "black"); #plot the histogram of the random numbers, with black edges
```

We can see that different values have been grouped, as we obtained 10 bins for 30 values. 
We can define the number of bins.

```python
plt.hist(group1, bins=30, edgecolor = "black"); #plot the histogram of the random numbers, with black edges and 30 bins
```

Like in normal distribution, the more values we have, the more the plot will resemble a bell. 

```python
rng_bell = np.random.default_rng(67890)
bell_group = rng.normal(50, 10, 3000) #generate 3000 random numbers with a normal distribution with mean 50 and standard deviation 10
plt.hist(bell_group, bins=30, edgecolor = "black"); #plot the histogram of the random numbers, with black edges and 30 bins
```

#### Create more groups for analysis:

```python
group2 = rng.normal(65, 12, 30) #generate 30 random numbers with a normal distribution with mean 65 and standard deviation 12
print(group2)
```

```python
group3 = rng.normal(85, 8, 30) #generate 30 random numbers with a normal distribution with mean 85 and standard deviation 8
print(group3)
```

We can create an image with the histogram plots for all the 3 groups.

We can use the alpha to make the plots 50% transparent.

```python
plt.hist(group1, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with blue color, black edges and 50% transparency.
plt.hist(group2, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with red color, black edges and 50% transparency.
plt.hist(group3, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with green color, black edges and 50% transparency.
legend_labels = ['group1', 'group2', 'group3'] #list of labels for the legend
plt.legend(legend_labels); #set the legend labels
```

Histograms are good to look at groups one at a time. Plots like the above one are messy.
