## Bar Charts

https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py

Bar charts are used for discrete type of data, not continuous.



### Example of a bar chart with individual bar colors

https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html#sphx-glr-gallery-lines-bars-and-markers-bar-colors-py

Plot fruits against their count, both discrete type of data.

X axis is discrete. The width of the bars is by default 0.8 (has no meaning) and bars are equally-spaced. 

The height of the bars, displayed on the y axis,  is instead important, as it indicates the count of the elements displayed on the x axis. 

```python
fig, ax = plt.subplots(); #create a figure and a set of subplots (even if we create just one plot now)

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange'] #list of labels of each bar. The underscore before 'red' is to make the label invisible.
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange'] #list of colors of each bar. tab: is a prefix to use the Tableau colors.

ax.bar(fruits, counts, label=bar_labels, color=bar_colors) #set the labels and colors of the bars

ax.set_ylabel('fruit supply') #set the y-axis label
ax.set_title('Fruit supply by kind and color') #set the title of the plot
ax.legend(title='Fruit color') #set the legend title
plt.show() #display the plot
```

```python
# simpler version of the above plot
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
plt.bar(fruits, counts);
```

## Histograms
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib.axes.Axes.hist

Histograms are used for continuous type of dat, where each value tends to be unique and trends can be observed. 




### Create data for analysis using NumPy Random Generator:

https://numpy.org/doc/stable/reference/random/generator.html

https://www.w3schools.com/python/numpy/numpy_random.asp

We want to generate a set of random normal numbers (normally distributed values).


```python
rng = np.random.default_rng(12345) #create a random number generator
group1 = rng.normal(75, 10, 30) #generate 30 random numbers with a normal distribution with mean 75 and standard deviation 10
print(group1)
```

Histograms can be used with continuous values and show trends, instead of single observations (like bar charts do).

```python
plt.hist(group1, edgecolor = "black"); #plot the histogram of the random numbers, with black edges
plt.title('Group 1') #set the title of the plot
```

We can see that different values have been grouped, as we obtained 10 bins for 30 values. 
We can define the number of bins.

```python
plt.hist(group1, bins=30, edgecolor = "black"); #plot the histogram of the random numbers, with black edges and 30 bins
plt.title('Group 1 with 30 bins') #set the title of the plot
```

Like in normal distribution, the more values we have, the more the plot will resemble a bell. 

```python
bell_group = rng.normal(50, 10,3000) #generate 3000 random numbers with a normal distribution with mean 50 and standard deviation 10
plt.hist(bell_group, bins=30, color = "red", edgecolor = "black"); #plot the histogram of the random numbers, with black edges and 30 bins
plt.title('Normal Distribution Plot') #set the title of the plot
```

#### Create more groups for analysis:

```python
group2 = rng.normal(65, 12, 30) #generate 30 random numbers with a normal distribution with mean 65 and standard deviation 12
print(group2)
```

```python
group3 = rng.normal(85, 8, 30) #generate 30 random numbers with a normal distribution with mean 85 and standard deviation 8
print(group3)
```

We can create an image with the histogram plots for all the 3 groups.

We can use the alpha to make the plots 50% transparent.

```python
plt.hist(group1, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with blue color, black edges and 50% transparency.
plt.hist(group2, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with red color, black edges and 50% transparency.
plt.hist(group3, edgecolor = "black", alpha = 0.5); #plot the histogram of the random numbers, with green color, black edges and 50% transparency.
legend_labels = ['group1', 'group2', 'group3'] #list of labels for the legend
plt.legend(legend_labels); #set the legend labels
plt.title('Multiple Groups Plot') #set the title of the plot
```

Histograms are good to look at groups one at a time. Plots like the above one are messy.

### Store Data

https://numpy.org/doc/stable/reference/generated/numpy.array.html

Create an array storing the data of the 3 groups.

```python
data = np.array([group1, group2, group3]) #create a 2D array with the three groups of random numbers
print(data)
```

```python
(data.shape) #check the shape of the array (3 rows and 30 columns)
```

### Descriptive Statistics

Mean: 

https://numpy.org/doc/stable/reference/generated/numpy.mean.html

Median:

https://numpy.org/doc/stable/reference/generated/numpy.median.html

Standard Deviation: 

https://numpy.org/doc/stable/reference/generated/numpy.std.html

```python
# calculate the mean of each group
means = np.mean(data, axis=1) #calculate the mean of each row   
print(f'The means are: {means}')
#calculate the standard deviation of each group
stds = np.std(data, axis=1) #calculate the standard deviation of each row
print(f'The standard deviations are: {stds}')
#Calculate the median of each group
medians = np.median(data, axis=1) #calculate the median of each row
print(f'The medians are: {medians}')
#calculate the maximum of each group
maxs = np.max(data, axis=1) #calculate the maximum of each row
print(f'The maximums are: {maxs}')
#calculate the minimum of each group
mins = np.min(data, axis=1) #calculate the minimum of each row
print(f'The minimums are: {mins}')
```

### Quartiles

Divide data into 4 equal quarters (25% each) and help to:

- identify outliers

- measure the central tendency (mean, median, mode)

- compare different data sets more easily

- identify skewness in data (asymmetric data distribution)


https://numpy.org/doc/stable/reference/generated/numpy.quantile.html

https://www.codecademy.com/learn/learn-statistics-with-python/modules/quartiles-quantiles-and-interquartile-range/cheatsheet

https://www.codecademy.com/learn/learn-statistics-with-python/modules/boxplots/cheatsheet



Data can have different types of distribution:

- Normal (Gaussian) Distribution

- Skewed Distribution

- Uniform Distribution

- Bimodal Distribution

- Multimodal Distribution

- Exponential Distribution

- Binomial Distribution

- Poisson Distribution

```python
# calculate the quartiles of each group
# Q1 
np.quantile(data, 0.25, axis=1) #calculate the 25th percentile of each row (Q1)
```

```python
# Q2 (median)
np.quantile(data, 0.5, axis=1) #calculate the 50th percentile of each row (Q2)
```

```python
# Q3
np.quantile(data, 0.75, axis=1) #calculate the 75th percentile of each row (Q3)
```

```python
# we can also calculate all the quartiles at once
np.quantile(data, [0.25, 0.5, 0.75], axis=1) #calculate the 25th, 50th and 75th percentiles of each
```

## Box Plots

Can be used to observe a single group but also to compare more than one group. 

Box Plots are very useful to observe the quartiles.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.boxplot.html

```python
# create a boxplot of the group1
fig, ax = plt.subplots()
ax.boxplot(group1);
```

```python
# create boxplots of all data
fig, ax = plt.subplots()
ax.boxplot(data.T); #transpose the data to have the groups in the x-axis and the values
```

Add title:

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html

Add labels:

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html

Set lables of X axis ticks:

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticks.html



```python
fig, ax = plt.subplots()
ax.boxplot(data.T, patch_artist= True); #transpose the data to have the groups in the x-axis and the values, and fill the boxes with colors
ax.set_title('Boxplot of the three groups', fontsize = 16) #set the title of the plot
ax.set_xlabel('Groups', fontsize = 14) #set the x-axis label
ax.set_ylabel('Values', fontsize = 14) #set the y-axis label
ax.set_xticklabels(['group1', 'group2', 'group3'], fontsize = 10) #set the x-axis labels
ax.grid(axis = "y", linestyle = "--", alpha = 0.5) #add a grid to the plot on the y-axis with dashed lines and 50% transparency
plt.show() #display the plot
```
