# **Scatterplots**

Scatterplots are used to compare two related variables.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

https://matplotlib.org/stable/gallery/shapes_and_collections/scatter.html

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

Import Diabetes dataset from sklearn datasets library.

https://scikit-learn.org/stable/datasets/toy_dataset.html

```python
dia = datasets.load_diabetes(scaled = False); # load the diabetes dataset
```

```python
dia # print the dataset
```

```python
dia.data[:10]  # print the data of the first 10 samples
```

```python
dia.data.shape # print the shape of the data
```

```python
dia.keys() # print the keys of the dataset
```

```python
dia.feature_names # print the feature names and check if they are the same as the column names in the data
```

```python
bp = dia.data[:,3] # access the 4th column of the data (blood pressure)
bp[:5] # print the first 5 rows of the blood pressure column
```

```python
bmi = dia.data[:, 2] # access the 3rd column of the data (BMI)
bmi[:5] # print the first 5 rows of the BMI column
```

Create the scatterplot to analyze the relationship between the blood pressure and the BMI:

```python
ax, fig = plt.subplots() # create a figure and axis object
fig.scatter(bmi, bp) # create a scatter plot of BMI vs blood pressure
fig.set_xlabel('BMI') # set the x-axis label
fig.set_ylabel('Blood Pressure') # set the y-axis label
fig.set_title('BMI vs Blood Pressure') # set the title of the plot
plt.show()
```

BMI and Blood Pressure don't look like totally unrelated, but don't show a strong relationship either.

### Improving the plot

Change the markers:

https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers


```python
# re-create the plot with a different marker
ax, fig = plt.subplots() 
fig.scatter(bmi, bp, marker = '.') # create a scatter plot of BMI vs blood pressure using smaller dots for better resolution
fig.set_xlabel('BMI')
fig.set_ylabel('Blood Pressure')
fig.set_title('BMI vs Blood Pressure') 
plt.show()
```

Change the color of the markers based on sex:

```python

sex = dia.data[:, 1] 
sex[:5] # print the first 5 rows of the sex column
```

```python
np.unique(sex) # print the unique values of sex column to make sure there are only two types of values
```

```python
ax, fig = plt.subplots() 
fig.scatter(bmi, bp, c = sex, marker = '.') # change the color of the dots based on sex
fig.set_xlabel('BMI')
fig.set_ylabel('Blood Pressure')
fig.set_title('BMI vs Blood Pressure') 
plt.show()
```

Change the color markers based on age:

```python
age = dia.data[:, 0]
age[:5] # print the first 5 rows of the age column
```

```python
np.unique(age) # print the unique values of age column to make sure there are only two types of values
```

```python
ax, fig = plt.subplots() 
fig.scatter(bmi, bp, c = age, marker = '.') # change the color of the dots based on sex
fig.set_xlabel('BMI')
fig.set_ylabel('Blood Pressure')
fig.set_title('BMI vs Blood Pressure') 
plt.show()
```

Change markers dimensions based on age:

```python
ax, fig = plt.subplots() 
fig.scatter(bmi, bp, c = sex, s = age, marker = '.') # change the size of the dots based on age
fig.set_xlabel('BMI')
fig.set_ylabel('Blood Pressure')
fig.set_title('BMI vs Blood Pressure') 
plt.show()
```
