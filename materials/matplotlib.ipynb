## X Squared Plot

```python
# create an array composed by 100 numbers using NumPy (domain of x values)
x = np.linspace (0.0, 10.0, 100)  #100 number of points

print (x)
```

```python
# create domain of y values
y = x ** 2

print (y)
```

```python
# plot using Pyplot

plt.plot (x, y);    # include semicolon not to print out the function
```

```python
# we can modify the style of the plot:
plt.plot (x, y, ".r");   # red dots 
```

```python
plt.plot ( x, y, "g-");
plt.xlabel("x")                 # include labels on the axis
plt.ylabel("$y = x^2 $")     # use math LaTex notation ("$ $") 

```

```python
# insert legend:
plt.plot (x, y, "y-", label = "$y = x^2$");   # add y legend on the top left corner
plt.xlabel('x')         # add axis labels
plt.ylabel('$x^2$')
plt.legend()
plt.title(" X Squared Plot")        # add title
```

## Comparing Plots

### Exponential Growth Plot

Plot 2 different functions and compare them.

```python
# domain of x values
x = np.linspace(0.0, 10.0, 101)

# create 2 domains of y values
y1 = x ** 2     # polinomial growth

y2 = 2 ** x    # exponential growth
```

```python
# create plot of x against y1
plt.plot(x, y1, "r-", label = "$y = x^2$")  # add y legend on the top left corner

# create plot of x against y2
plt.plot(x, y2, "g-", label = "$y = 2^x$")  # add y legend on the top left corner

plt.legend()  # add legend

plt.xlabel("x")  # add x label
plt.ylabel("y")  # add y label

plt.title("Comparison of two Exponential Growth Plots")  # add title
```

Exponential functions plots always grow much faster than the polinomial ones. 

## Integers

### Scatter Plot

np.linspace() produces arrays of floats, that cannot be integers. 

https://numpy.org/doc/stable/reference/generated/numpy.linspace.html


```python
#
z = np.linspace(1, 100, 100)
z
```

```python
print(z.dtype)
```

To create an array of integers with np.linspace() we need to specify the non positional argument dtype.

If the number of values requires the creation of the floats, np.linspace() rounds numbers to floor.

```python
zz = np.linspace(1, 100, 100, dtype = int)
zz
```

```python
print(zz.dtype)
```

To create an array of integers we can use np.arange().

https://numpy.org/doc/stable/reference/generated/numpy.arange.html

```python
# create a domain of x values
x = np.arange(1, 101, 1)
x
```

```python
print(x.dtype)
```

```python
# create domain of y values
y = x**3 + 2 * x**2 + 1
print(y.dtype)
```

The dtype of y is also int64, because it infers it from x dtype. 

If x dtype was a float, also y dtype would have been a float.

```python
# create scatter plot of x against y
plt.plot(x, y, "b.", label = "$y = x^3 + 2x^2 + 1$")  # add y legend on the top left corner
plt.xlabel("x")  # add x label
plt.ylabel("y")  # add y label
plt.legend()  # add legend
plt.title("Plotting Integers")  # add title
```
